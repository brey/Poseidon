## Verif examples - Notebook

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
import verif
```

```python
data=verif.Data.Data(['/DATA/critechuser/REPOS/verif/examples/kf.txt','/DATA/critechuser/REPOS/verif/examples/raw.txt'])
```

```python
data.getFilenames()
```

```python
taylor=verif.Output.Taylor()
```

```python
taylor.setFigsize((10,8))
```

```python
taylor.plot(data)
```

```python
error=verif.Output.Error()
```

```python
error.setFigsize((10,8))
```

```python
error.plot(data)
```

```python
reliability=verif.Output.Reliability()
```

```python
reliability.setThresholds([0])
```

```python
reliability.setFigsize((10,8))
```

```python
reliability.plot(data)
```

```python
m=verif.Metric.Pit('pit')
```

```python
pithist=verif.Output.PitHist(m)
```

```python
pithist.setFigsize((10,8))
```

```python
pithist.plot(data)
```
