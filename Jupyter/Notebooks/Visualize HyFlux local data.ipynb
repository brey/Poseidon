Read data

```python
pfiles=glob.glob(PATH+'*_H_*.tif')
```

```python
pfiles.sort()
```

```python
dat=getmap(pfiles[0])
```

```python
gt=dat.GeoTr

width=dat.NCOLS
height=dat.NROWS

minx = gt[0]
miny = gt[3] + width*gt[4] + height*gt[5]
maxx = gt[0] + width*gt[1] + height*gt[2]
maxy = gt[3]

lon=np.linspace(minx,maxx,width,endpoint=True)
lat=np.linspace(miny,maxy,height,endpoint=True)
```

```python
minlat=13.5
maxlat=14.5
minlon=-61.5
maxlon=-60.5
```

```python
m = Basemap(projection='cyl',llcrnrlat=minlat,urcrnrlat=maxlat,\
             llcrnrlon=minlon,urcrnrlon=maxlon,resolution='c')
```

```python
h=dat.data
fill_value=dat.nan
iframes=np.size(pfiles)
fig = plt.figure()
bh=np.ma.masked_array(h,h==fill_value)
H=m.imshow(bh,animated=True)    
tframe=pfiles[0].split('_')[-1].split('.')[0]
cb=plt.colorbar(H)
fig.text(.01,.95,'frame= {} {} {} {}'.format(tframe[:4],tframe[4:6],tframe[6:8],tframe[8:]),horizontalalignment='left',size='small')
```

```python
h=dat.data
fill_value=dat.nan
iframes=np.size(pfiles)
fig = plt.figure()
bh=np.ma.masked_array(h,h==fill_value)
H=m.imshow(np.flipud(bh),animated=True)    
tframe=pfiles[0].split('_')[-1].split('.')[0]
m.drawcoastlines()
cb=plt.colorbar(H)
fig.text(.01,.95,'frame= {} {} {} {}'.format(tframe[:4],tframe[4:6],tframe[6:8],tframe[8:]),horizontalalignment='left',size='small')
```

```python
def animate(i):
    
    dat=getmap(pfiles[i])
    h=dat.data
    bh=np.ma.masked_array(h,h==fill_value)
    fig.texts.pop()
    H.set_array(np.flipud(bh))
    m.drawcoastlines()
    tframe=pfiles[i].split('_')[-1].split('.')[0]
    fig.text(.01,.95,'frame= {} {} {} {}'.format(tframe[:4],tframe[4:6],tframe[6:8],tframe[8:]),horizontalalignment='left',size='small')
```

```python
# call the animator.  blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, frames=iframes, interval=200, blit=False, repeat=False)
```
