```python
from matplotlib import animation
import matplotlib.pyplot as plt

import glob
import numpy as np
from pmap import getmap
```

```python
matplotlib.get_backend()
```

```python
from tempfile import NamedTemporaryFile

VIDEO_TAG = """<video controls>
 <source src="data:video/x-webm;base64,{0}" type="video/webm">
 Your browser does not support the video tag.
</video>"""

def anim_to_html(anim):
    if not hasattr(anim, '_encoded_video'):
        with NamedTemporaryFile(suffix='.webm') as f:
            anim.save(f.name, fps=6, extra_args=['-vcodec', 'libvpx'])
            video = open(f.name, "rb").read()
        anim._encoded_video = video.encode("base64")
    
    return VIDEO_TAG.format(anim._encoded_video)
```

```python
from IPython.display import HTML

def display_animation(anim):
    plt.close(anim._fig)
    return HTML(anim_to_html(anim))
```

```python
PATH='/mnt/Tsunamiweb/processed/2015/FIX_MED_SEA/tif_20150204.00/'
```

Read data

```python
pfiles=glob.glob(PATH+'VMAX*.tif')
```

```python

dat=getmap(pfiles[0])

h=dat.data
fill_value=dat.nan
iframes=np.size(pfiles)
fig = plt.figure()
bh=np.ma.masked_array(h,h==fill_value)
H=plt.imshow(bh,animated=True)
cb=plt.colorbar(H)
fig.text(.01,.95,'frame= '+np.str(0),horizontalalignment='left',size='small')
```

```python
def animate(i):
    
    dat=getmap(pfiles[i])
    h=dat.data
    bh=np.ma.masked_array(h,h==fill_value)
    fig.texts.pop()
    H.set_array(bh)
    fig.text(.01,.95,'frame= '+np.str(i),horizontalalignment='left',size='small')
```

```python
# call the animator.  blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, frames=iframes, interval=200, blit=False)

# call our new function to display the animation
display_animation(anim)
```
