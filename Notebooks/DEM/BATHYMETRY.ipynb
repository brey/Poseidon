# Visualize & compare Dems

```python
import numpy as np
from netCDF4 import Dataset
from pyPoseidon.dem import *
from pyPoseidon.utils import *
```

```python
%matplotlib inline
```

```python
# Lat/Lon window
dic={'lon0':-35.99, # lat/lon window
     'lon1':42.99,
     'lat0':25,
     'lat1':85.}
```

```python
# Gebco 14
g14 = gebco14(dpath='/home/critechproc/STORAGE/BATHYMETRY/GLOBAL/GEBCO_2014_2D.nc',**dic)
sea14 = np.ma.masked_array(g14.val,g14.val>0.) #mask land
```

```python
plt.figure(figsize=(14,8))
plt.contourf(g14.dlons,g14.dlat,sea14)
plt.colorbar()
```

```python
# Gebco 08
g08 = gebco08(dpath='/home/critechproc/STORAGE/BATHYMETRY/GLOBAL/gebco30_DELTARES.nc',**dic)
sea08 = np.ma.masked_array(g08.val,g08.val>0.) #mask land
```

```python
plt.figure(figsize=(14,8))
plt.contourf(g08.dlons,g08.dlat,sea08)
plt.colorbar()
```

```python
#emodnet
edt=Dataset('/home/critechproc/STORAGE/BATHYMETRY/EMODNET/TEMP_bathymetry.nc')
```

```python
lat = edt.variables['latitude'][:]
lon = edt.variables['longitude'][:]
depth = edt.variables['depth'][:]
```

```python
x,y=np.meshgrid(lon,lat)
```

```python
sea = np.ma.masked_array(depth,depth>0)
```

```python
plt.figure(figsize=(14,8))
plt.contourf(x,y,sea)
plt.colorbar()
```
